````python
tags = ['h1', 'p']
text = [element.text for element in soup.find_all(tags)][2:]

name_list = []

for item in text: 
    x = item.strip()
    name_list.append(x)
name_list
````

````python
def names(l, n):
    for i in range(0, len(l), n):
        yield l[i:i+n]

name_list1 = names(name_list, 3) 
name_list2 = list(name_list1)
name_list2
````

````python
trending_devs = []

for i in range(len(name_list2)):
        element = name_list2[i][0] + ' (' + name_list2[i][1] + ')'  
        trending_devs.append(element)

trending_devs
````

#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page.
